```python
book1=book.rename({'User.ID':'Userid','Book.Title':'Booktitle','Book.Rating':'Bookrating'},axis=1)
```

```python
book1
```

```python
#number of unique users in the dataset
book1.Userid.nunique()
```

```python
len(book1.Bookrating.unique())
```

```python
#user_book1 = book1.pivot(index=index,columns=columns,values=values).reset_index(drop=True)
user_book1 = book1.pivot_table(index='Userid',
                         columns='Booktitle',
                         values='Bookrating')
```

```python
book1.Userid.unique()
```

```python
user_book1.index=book1.Userid.unique()
```

```python
user_book1
```

```python
#Impute those NaNs with 0 values
user_book1.fillna(0, inplace=True)
```

```python
user_book1
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book1.values,metric='cosine')
user_sim
```

```python
user_sim.shape
```

```python
len(user_sim)
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids
user_sim_df.index = book1.Userid.unique()
user_sim_df.columns = book1.Userid.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:10]
```
