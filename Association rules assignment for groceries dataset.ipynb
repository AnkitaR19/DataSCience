# Pre-Processing

```python
df=pd.get_dummies(groc)# this creates dummy variables for all the data
df.head()
#te = TransactionEncoder()
#te_ary = te.fit(groc).transform(groc)
#df = pd.DataFrame(te_ary, columns=te.columns_)
```

# Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
#########import numpy as np
###################np.vectorize(frequent_itemsets)
```

```python
 ####################fv1 = np.vectorize(frequent_itemsets, otypes=[int])
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
len(rules)
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>=1]
#this gives rules greater than 1
```
