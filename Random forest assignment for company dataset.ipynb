```python
# AdaBoost Classification
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
cp=pd.read_csv("E:/ASSIGNMENT FROM EXCELR/Random Forests/Company_Data (1).csv")
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
cp["ShelveLoc"] = label_encoder.fit_transform(cp["ShelveLoc"])
cp["Urban"] = label_encoder.fit_transform(cp["Urban"])
cp["US"] = label_encoder.fit_transform(cp["US"])
array = cp.values
X = array[:,0:10]
Y = array[:,10]
num_trees = 50
seed=7
kfold = KFold(n_splits=5, random_state=seed)
model = AdaBoostClassifier(n_estimators=num_trees, random_state=seed)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# Stacking Ensemble for Classification
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
#cp=pd.read_csv("E:/ASSIGNMENT FROM EXCELR/Random Forests/Company_Data (1).csv")
from sklearn import preprocessing
#label_encoder = preprocessing.LabelEncoder()
#cp["ShelveLoc"] = label_encoder.fit_transform(cp["ShelveLoc"])
#cp["Urban"] = label_encoder.fit_transform(cp["Urban"])
#cp["US"] = label_encoder.fit_transform(cp["US"])
#array = cp.values
#X = array[:,0:10]
#Y = array[:10]
kfold = KFold(n_splits=6, random_state=7)
# create the sub models
estimators = []
model1 = LogisticRegression(max_iter=500)
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))
model3 = SVC()
estimators.append(('svm', model3))
# create the ensemble model
ensemble = VotingClassifier(estimators)
results = cross_val_score(ensemble, X, Y, cv=kfold)
print(results.mean())
```
